```python
#analytical solution
n = 200
q = 1500
EI = 600
L = 3
x = np.linspace(L,0,n)
y = (((-q/(24*EI))*((x**4)-(2*L*x**3))) + ((x*-q*(L**3))/(24 * EI)))
plt.xlabel("x (meters)")
plt.ylabel("y (meters)")
plt.title("PDE Analytic Solution - Beam Deflecton")
plt.plot(x,y,label='$f(x) = -5/4 ((x^4/12)-(3x^3/6)+(5*27/48))$')
plt.savefig("PDE Analytic")
```
